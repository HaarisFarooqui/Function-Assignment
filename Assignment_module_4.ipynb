Q1  What is the difference between a function and a method in Python?
Ans- A **function** is an independent block of code defined using the `def` keyword, while a **method** is a function associated with an object and called using the **dot notation** (`object.method()`). Methods belong to classes, whereas functions can exist outside of classes.

Q2 - Explain the concept of function arguments and parameters in Python
Ans - In Python, **parameters** are variables defined in a function's declaration (e.g., `def my_function(param):`), while **arguments** are the actual values passed to the function when it's called (e.g., `my_function(5)`). Parameters act as placeholders; arguments are the data they receive.


Q3- What are the different ways to define and call a function in Python?
Ans -Here are the main ways to define and call a function in Python:

1. **Basic Definition**: `def func():` - No parameters, called as `func()`.
2. **With Parameters**: `def func(x, y):` - Called with arguments, e.g., `func(1, 2)`.
3. **Default Parameters**: `def func(x, y=10):` - Called as `func(5)` or `func(5, 20)`.
4. **Keyword Arguments**: `def func(x, y):` - Called as `func(y=2, x=1)`.
5. **Arbitrary Args**: `def func(*args):` - Called with any number of args, e.g., `func(1, 2, 3)`.
6. **Arbitrary Keyword Args**: `def func(**kwargs):` - Called as `func(a=1, b=2)`.
7. **Lambda Function**: `func = lambda x: x + 1` - Called as `func(5)`.

Each is defined with `def` (except lambda) and called by name with parentheses.

Q4- What is the purpose of the `return` statement in a Python function?
Ans - The `return` statement in a Python function specifies the value the function sends back to the caller. It ends the function's execution and optionally provides output (e.g., `return 5` or `return x + y`). Without `return`, the function returns `None` by default.

Q5- What are iterators in Python and how do they differ from iterables?
Ans - In Python, an **iterable** is any object that can be looped over (e.g., lists, strings), capable of returning an iterator via `iter()`. An **iterator** is an object that produces one item at a time using `next()`, maintaining its state. All iterators are iterables, but not all iterables are iterators (e.g., a list is iterable but not an iterator until `iter(list)` is called).

Q6- Explain the concept of generators in Python and how they are defined
Ans - Generators in Python are a type of iterator that yield values one at a time, pausing execution between yields to save memory. They’re defined using a function with the `yield` keyword (e.g., `def gen(): yield 1; yield 2`) or as a generator expression (e.g., `(x * 2 for x in range(3))`). Unlike regular functions, they don’t return all values at once but generate them on demand.

Q7- What are the advantages of using generators over regular functions?
Ans - Generators have these advantages over regular functions:

1. **Memory Efficiency**: They yield one item at a time, unlike regular functions that store all values in memory.
2. **Lazy Evaluation**: Values are computed only when needed, improving performance.
3. **Infinite Sequences**: They can represent endless data streams (e.g., infinite numbers) without crashing.
4. **Simpler Code**: They avoid complex state management for iteration, using `yield` instead of manual tracking.

Q8-  What is a lambda function in Python and when is it typically used?
Ans - A **lambda function** in Python is a small, anonymous function defined with `lambda` (e.g., `lambda x: x + 1`). It can take any number of arguments but has a single expression. It’s typically used for short, one-off operations, like in `map()`, `filter()`, or sorting (e.g., `sorted(list, key=lambda x: x[1])`), where a full function definition isn’t needed.

Q9-  Explain the purpose and usage of the `map()` function in Python.
Ans- The `map()` function in Python applies a given function to each item in an iterable (e.g., list) and returns an iterator of the results. Syntax: `map(function, iterable)`. For example, `map(lambda x: x*2, [1, 2, 3])` yields `2, 4, 6`. It’s used for transforming data efficiently without explicit loops.

Q10- Here’s the difference in short:

- **`map()`**: Applies a function to each item in an iterable, returning an iterator of transformed values (e.g., `map(lambda x: x*2, [1, 2, 3])` → `2, 4, 6`).
- **`filter()`**: Applies a function to each item, returning an iterator of items where the function returns `True` (e.g., `filter(lambda x: x > 1, [0, 1, 2])` → `2`).
- **`reduce()`**: From `functools`, applies a function cumulatively to items in an iterable, reducing it to a single value (e.g., `reduce(lambda x, y: x+y, [1, 2, 3])` → `6`).

**Key**: `map` transforms, `filter` selects, `reduce` aggregates.
